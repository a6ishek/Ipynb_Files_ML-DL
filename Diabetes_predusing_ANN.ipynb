```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```
import tensorflow.keras as tf
```

```
Data = pd.read_csv("/content/diabetes.csv")
Data
```

```
Data.columns
```

```
Data.info()
```

```
X = Data.iloc[:,:8]
Y = Data.iloc[:,-1]
print(X,Y)
np.shape(X)
```

```
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.15,random_state = 1)
```

```
Y.value_counts()
```

```
def ratio(x,y):#to find x's percent in overall data(x+y)
   return (x/(x+y))*100

```

```
ratio(425,227)
```

```
Y_train.value_counts()
```

```
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values = 0,strategy='mean')
X_train = si.fit_transform(X_train)
X_test = si.fit_transform(X_test)
```

```
model = tf.models.Sequential()
model.add(tf.layers.Dense(64,input_dim=8,activation='relu'))
model.add(tf.layers.Dense(128,activation='relu'))
model.add(tf.layers.Dense(256,activation='relu'))
model.add(tf.layers.Dense(1,activation='sigmoid'))
```

```
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```
model.fit(X_train,Y_train,epochs=500)
```

```
Y_pred = model.predict(X_test)
Y_pred
```

```
Y_pred = Y_pred.round()
```

```
Y_pred
```

```
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```
